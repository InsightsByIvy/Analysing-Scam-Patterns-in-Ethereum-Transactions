### Starting the cleaning process

1. Strip whitespace and lower case all column names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

2. Drop unnecessary columns

```python
df.drop(columns=['unnamed:_0', 'index'], inplace=True)
df.columns
```

Checking if the two columns ('erc20_uniq_sent_addr' and 'erc20_uniq_sent_addr.1') are identical.

```python
(df['erc20_uniq_sent_addr'] == df['erc20_uniq_sent_addr.1']).all()
```

Checking how much they differ.

```python
diff_count = (df['erc20_uniq_sent_addr'] != df['erc20_uniq_sent_addr.1']).sum()
print(f"Number of rows where values differ: {diff_count}")
```

Checking the values in both columns where the value is different.

```python
df_diff = df[df['erc20_uniq_sent_addr'] != df['erc20_uniq_sent_addr.1']]
df_diff[['erc20_uniq_sent_addr', 'erc20_uniq_sent_addr.1']].head(10)
```

Since the result shows zeros only, checking the unique values in column 'erc20_uniq_sent_addr.1'.

```python
df['erc20_uniq_sent_addr.1'].unique()
```

Visually assessing whether the two columns match, differ, or conflict.

```python
mask = (df['erc20_uniq_sent_addr.1'] != 0) & (df['erc20_uniq_sent_addr.1'].notna())
df.loc[mask, ['erc20_uniq_sent_addr', 'erc20_uniq_sent_addr.1']].head(10)
```

Renaming the column 'erc20_uniq_sent_addr.1' for clarity as it has unique values

```python
df = df.rename(columns={'erc20_uniq_sent_addr.1': 'erc20_uniq_sent_addr_alt'})
```

Since these columns ('ERC20_uniq_sent_token_name' and 'ERC20_uniq_rec_token_name') are showing as float and not string, checking if those float columns contain actual token names or just NaNs and numbers.

```python
print(df['erc20_uniq_sent_token_name'].dropna().unique()[:10])
print(df['erc20_uniq_rec_token_name'].dropna().unique()[:10])
```

Since I got numberical values back, likely the column name is misleading, therefore, going to fill the missing values and then rename the columns.

```python
df['erc20_uniq_sent_token_name'] = df['erc20_uniq_sent_token_name'].fillna(0).astype(int)
df['erc20_uniq_rec_token_name'] = df['erc20_uniq_rec_token_name'].fillna(0).astype(int)
```

For token type columns (erc20_most_sent_token_type, erc20_most_rec_token_type) filling missing value with a placeholder string like 'No_Token' or 'None':

```python
df['erc20_most_sen_toke_type'] = df['erc20_most_sent_token_type'].fillna('No_Token')
df['erc20_most_rec_token_type'] = df['erc20_most_rec_token_type'].fillna('No_Token')
```

```python
df.rename(columns={
    'erc20_uniq_sent_token_name': 'erc20_uniq_sent_token_count',
    'erc20_uniq_rec_token_name': 'erc20_uniq_rec_token_count'}, inplace=True)
```
